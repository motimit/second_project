# 팀원별 분석적 특성 비교

| 팀원 | 분석적 특성 | 강점 | 주요 초점 |
|------|-------------|------|-----------|
| **이찬희** | 구조적, 체계적 접근 | - 단계별 EDA 수행 (결측치 → 이상치 → 변수 탐색) <br> - 결측치 처리 기준 명확 (0~2% → drop, 70% 이상 → 삭제 후보) <br> - 수치형/범주형 변수 구분, 스케일링 후보 선정 | 데이터 품질 확인, 이상치 탐색, 분석 기반 마련 |
| **한규리** | 도메인 기반 해석, 정책/법적 요인 고려 | - 뉴욕주 다세대 주택법과 데이터 왜곡 사례 분석 (30일 이상 숙박) <br> - 리뷰 점수·리뷰 수 관계 확인 | 데이터가 현실 세계와 연결되는 방식 이해, 도메인 기반 인사이트 도출 |
| **김소희** | 목표 중심, 프로젝트 산출물 구조화 | - 프로젝트 정의, 분석 목표, 산출물/활용 방안 구조화 <br> - 데이터 구조·컬럼 타입 확인, 로그 변환 등 전처리 계획 <br> - 변수별 분포 및 시각화 강조 | 분석 과정 전체 계획, 가격 예측 모델링, 타겟 숙소 리스트 생성 |
| **윤혜빈** | 시장 구조 + 통계적 해석 | - Airbnb 시장 구조 분석: 공급 과잉, 경쟁 강도, 가격 편차 <br> - 변수별 가격 영향력 분석 (리뷰 수, 가용일수, 숙소 유형) <br> - 통계 지표 활용: 평균, 표준편차, 이상치 | 가격 결정 요인 이해, 최적 가격 전략 도출, 변수와 가격 관계 분석 |
