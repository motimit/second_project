### 📊 Airbnb NYC Listings 컬럼 정리 (현재 데이터셋 기준)

##### 🆔 1. 기본 식별 / 데이터 수집 정보   4ea
| 컬럼명 | 설명 |
|--------|------|
| `Unnamed: 0` | CSV 저장 과정에서 생성된 인덱스 컬럼 (분석 시 삭제 가능) |
| `id` | Airbnb에서 부여한 해당 숙소의 고유 식별자 |
| `source` | 데이터 수집 방식 ("neighbourhood search" 또는 "previous scrape") |
| `calendar_last_scraped` | 캘린더 정보가 마지막으로 수집된 날짜 |

##### 🏠 2. 숙소 기본 정보   11ea
| 컬럼명 | 설명 |
|--------|------|

| `description` | 숙소 상세 설명 |
| `neighborhood_overview` | 호스트가 작성한 동네 소개 |

| `accommodates` | 최대 수용 가능 인원 |
| `bathrooms` | 욕실 개수 |
| `bathrooms_text` | 욕실 수에 대한 텍스트 설명 |
| `bedrooms` | 침실 개수 |
| `beds` | 침대 개수 |
| `amenities` | 제공 편의시설 목록 (JSON 형식) |
